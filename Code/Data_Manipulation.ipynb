```
data1 = "/gdrive/My Drive/Final-Project---Netflix-Ratings/data/combined_data_1.txt"
data2 = "/gdrive/My Drive/Final-Project---Netflix-Ratings/data/combined_data_2.txt"
data3 = "/gdrive/My Drive/Final-Project---Netflix-Ratings/data/combined_data_3.txt"
data4 = "/gdrive/My Drive/Final-Project---Netflix-Ratings/data/combined_data_4.txt"
```

```
df1 = pd.read_csv(data1, header = None, names = ['Cust_Id', 'Rating'], usecols = [0,1])

df1['Rating'] = df1['Rating'].astype(float)

print('Dataset 1 shape: {}'.format(df1.shape))
print('-Dataset examples-')
print(df1.iloc[::5000000, :])
```

```
df2 = pd.read_csv(data2, header = None, names = ['Cust_Id', 'Rating'], usecols = [0,1])
df3 = pd.read_csv(data3, header = None, names = ['Cust_Id', 'Rating'], usecols = [0,1])
df4 = pd.read_csv(data4, header = None, names = ['Cust_Id', 'Rating'], usecols = [0,1])


df2['Rating'] = df2['Rating'].astype(float)
df3['Rating'] = df3['Rating'].astype(float)
df4['Rating'] = df4['Rating'].astype(float)

print('Dataset 2 shape: {}'.format(df2.shape))
print('Dataset 3 shape: {}'.format(df3.shape))
print('Dataset 4 shape: {}'.format(df4.shape))
```

```
df = df1

df.index = np.arange(0,len(df))
print('Full dataset shape: {}'.format(df.shape))
print('-Dataset examples-')
print(df.iloc[::5000000, :])

df_nan = pd.DataFrame(pd.isnull(df.Rating))
df_nan = df_nan[df_nan['Rating'] == True]
df_nan = df_nan.reset_index()

movie_np = []
movie_id = 1

for i,j in zip(df_nan['index'][1:],df_nan['index'][:-1]):
    # numpy approach
    temp = np.full((1,i-j-1), movie_id)
    movie_np = np.append(movie_np, temp)
    movie_id += 1

# Account for last record and corresponding length
# numpy approach
last_record = np.full((1,len(df) - df_nan.iloc[-1, 0] - 1),movie_id)
movie_np = np.append(movie_np, last_record)

print('Movie numpy: {}'.format(movie_np))
print('Length: {}'.format(len(movie_np)))

df = df[pd.notnull(df['Rating'])]

df['Movie_Id'] = movie_np.astype(int)
df['Cust_Id'] = df['Cust_Id'].astype(int)
print('-Dataset examples-')
print(df.iloc[::5000000, :])
df.to_csv("/gdrive/My Drive/Final-Project---Netflix-Ratings/data/output/full1.txt",index=False)
```
